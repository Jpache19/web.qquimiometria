```python
#Librarías
import numpy as np
import matplotlib.pyplot as plt
from math import exp
```

```python
#Condiciones del problema
m= 68.1
print("masa (kg):", m)
g= 9.81
print("gravedad (m/s):", g)
t= 10
print("tiempo (s):", t)
V= 40
print("velocidad (m/s) a los 10s:", V)
```

```python
#Ecuación de la constante de proporcionalidad "k" por métodos gráficos e iterativos
def x(k):
  return (((m*g/t)/k)*(1-np.exp((-t/m)*k))-V/t)
k= np.linspace(1,15,150)
```

```python
#graficación
plt.plot(k,x(k), color= "m")
plt. xlabel("k")
plt. ylabel ("x(k)")
plt. title("coeficiente de resistencia")
```
